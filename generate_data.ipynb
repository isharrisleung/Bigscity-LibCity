```python
import numpy as np
import pandas as pd
data_name = "jinan"

original_data = "./raw_raw_data/HGCN_data/"+data_name+"/original_data.npz"
sensor_ids = "./raw_raw_data/HGCN_data/"+data_name+"/graph_sensor_ids.txt"
distances = "./raw_raw_data/HGCN_data/"+data_name+"/distance.csv"

output_dir = "./raw_data/" + data_name + "/"

data = np.load(original_data)["data"]
sensor_ids = np.loadtxt(sensor_ids,delimiter=',')
sensor_ids=np.int32(sensor_ids)

distance_df = pd.read_csv(distances, dtype={'from': 'int', 'to': 'int'})
```

```python
# geo
sensor_ids
geo_csv = pd.DataFrame(sensor_ids, columns=["geo_id"])
geo_csv["type"] = "Point"
geo_csv["coordinates"] = None
geo_csv.to_csv(output_dir+data_name+".geo", index=False)
geo_csv
```

```python
# rel
distance_df.columns = ["origin_id", "destination_id", "cost"]
tmp = list(range(len(distance_df)))
tmp_df = pd.DataFrame(tmp, columns=["rel_id"])
tmp_df["type"] = "geo"
distance_df = pd.concat([tmp_df, distance_df], axis=1 )
distance_df.to_csv(output_dir+data_name+".rel", index=False)
distance_df
```

```python
# dyna
print(data.shape)
# print(data[0])
# print([geo_csv["geo_id"][0]] * 10)
gg = list(geo_csv["geo_id"])
# print(gg)
sensor, timestep, _ = data.shape
data = data.reshape(sensor, timestep)
# print(data.shape)
entity_id = []
traffic_speed = data.reshape(sensor*timestep,)
# print(traffic_speed.shape)
for i in range(sensor):
    entity_id.extend([gg[i]] * timestep)
# for i in range(len(data)):
#     entity_id.extend([geo_csv["geo_id"][i]] * len(data[i]))
#     traffic_speed.extend(data[i])
# print(entity_id)
dyna_df = list(range(len(traffic_speed)))
# print(len(dyna))
dyna_df = pd.DataFrame(dyna_df, columns=["dyna_id"])
dyna_df["type"] = "state"
dyna_df["time"] = None
dyna_df["entity_id"] = entity_id
dyna_df["traffic_speed"] = traffic_speed
dyna_df.to_csv(output_dir+data_name+".dyna", index=False)
dyna_df
```
